4.Discuss the different types of operators in Python and provide examples of how they are used.
- Python provides a variety of operators that allow you to perform operations on variables and values. Here's a breakdown of the different types of operators in Python and examples of how they are used:
  - Arithmetic Operators
    - These operators are used to perform mathematical operations.
     - x = 5
     -  y = 3
     -  print(x + y)   # Output: 8
     -  print(x * y)   # Output: 15
     -  print(x ** y)  # Output: 125

    - hese operators are used to compare two values and return a Boolean result
     - x = 5
     - y = 3
     - print(x == y)  # Output: False
     - print(x > y)   # Output: True
  - Logical Operators
    - These operators are used to combine conditional statements.
     - x = 5
     - y = 3
     - print(x > y and y > 2)  # Output: True
     - print(not (x < y))      # Output: True

    - These operators are used to assign values to variables.
     - x = 10
     - x += 5  # Now x = 15
     - x *= 2  # Now x = 30
     - print(x)  # Output: 30
  - Identity Operators
    - These operators are used to compare the memory locations of two objects.
     - x = [1, 2, 3]
     -  y = [1, 2, 3]
     - z = x
     - print(x is y)    
     - Output: False
     -  print(x is z)

5.Explain the concept of type casting in Python with examples.
  - Type casting (or type conversion) is the process of converting a variable from one data type to another. In Python, you can explicitly convert between types using type casting functions, or Python can implicitly convert types when necessary.
    - Implicit Type Casting:
      - Implicit Type Casting (also known as automatic type conversion) happens automatically in Python when the interpreter converts a smaller or less precise data type to a larger or more precise one to prevent data loss.
       - x = 5       
       - y = 3.2    
       - result = x + y
       - print(result)   
       - Output: 8.2
    - Explicit Type Casting (Manual Conversion):
      -  Converting a float to an integer
      - x = 4.7
      - y = int(x)  
      - print(y)     
      - Output: 4

      - Converting a string to an integer
      - s = "123"
      - z = int(s)  
      - print(z)     
      - Output: 123

7.Describe the different types of loops in Python and their use cases with examples.
- Loops in Python allow you to execute a block of code repeatedly as long as a condition is true. Python provides two main types of loops:
   - for loop
    - The for loop in Python is typically used to iterate over sequences such as lists, tuples, dictionaries, sets, or strings. It iterates over each item in the sequence and executes the code block for each item.
    - example 1:
   
     - fruits = ["apple", "banana", "cherry"]
     - for fruit in fruits:
     -     print(fruit)
     - output:
     - apple
     - banana
     - cherry
    - example 2:

     - for i in range(5):
     -     print(i)
     - output
     - 0
     - 1
     - 2
     - 3
     - 4
  - while loop
   - The while loop is used to repeat a block of code as long as a condition is true. It is generally used when you don't know the exact number of iterations in advance but know the condition that must hold for the loop to continue.
    - Example
    - i = 0
    - while i < 5:
    -  print(i)
    -   i += 1
    - output:
    - 0
    - 1
    - 2
    - 3
    - 4
